```python
import numpy as np
import tensorflow as tf
import tensorflow.keras.datasets as ds

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
(x_train,y_train),(x_test,y_test)=ds.mnist.load_data()
x_train=x_train.reshape(60000,784)
x_test=x_test.reshape(10000,784)
x_train=x_train.astype(np.float32)/255.0
x_test=x_test.astype(np.float32)/255.0
y_train=tf.keras.utils.to_categorical(y_train,10)
y_test=tf.keras.utils.to_categorical(y_test,10)
```

```python
#0.1

mlp=Sequential()
mlp.add(Dense(units=512,activation='tanh',input_shape=(784,)))
mlp.add(Dense(units=10,activation='softmax'))

mlp.compile(loss='MSE',optimizer=Adam(learning_rate=0.1),metrics=['accuracy'])
mlp.fit(x_train,y_train,batch_size=128,epochs=50,validation_data=(x_test,y_test),verbose=2)

res_0=mlp.evaluate(x_test,y_test,verbose=0)
print('정확률=',res_0[1]*100)
```

```python
# 0.01
mlp2=Sequential()
mlp2.add(Dense(units=512,activation='tanh',input_shape=(784,)))
mlp2.add(Dense(units=10,activation='softmax'))

mlp2.compile(loss='MSE',optimizer=Adam(learning_rate=0.01),metrics=['accuracy'])
mlp2.fit(x_train,y_train,batch_size=128,epochs=50,validation_data=(x_test,y_test),verbose=2)

res_1=mlp2.evaluate(x_test,y_test,verbose=0)
print('정확률=',res_1[1]*100)
```
